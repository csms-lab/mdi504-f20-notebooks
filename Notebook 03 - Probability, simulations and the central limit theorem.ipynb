# Probability basics
---

First, we recall some basic definitions. Formally, [**random variable**](https://en.wikipedia.org/wiki/Random_variable) $X$ is a mapping  from a sample space to the real numbers, but informally you can think of a random variable as an outcome of an experiment and instead of being assigned a single number, we assign it a **probability distribution**. To be circular, a random variable is random or non-deterministic.

The probability distribution describes:

a) (Frequentist) A normalized histogram of the values $X$ takes if you run the experiment a large number of times or

b) (Bayesian) Your beliefs about what the unknown value of $X$ can take if you were to run an experiment.


$X$ can be a **discrete random variable** - the values it can take on are finite, or at least countable (like the integers) -- or a **continuous random variable** -- the values it can take are a non-countable subset of the real numbers. The flip of a coin or a roll of a die are discrete random variables. The strength of a material can be modeled as a continuous random variable.

## Distributions

The probability distribution of an RV is encoded in its probability function. For discrete RVs, we talk of its **probability mass function** (pmf) which just describes the  probability the RV $X$ takes on any particular value. In symbols, we write:

$$ p(x) = \mathbb P[X = x].$$

To ask what the probability is that $X$ is in some set $\mathcal X$, we can just sum up the probabilities
$$ \mathbb P[X \in \mathcal X] = \sum_{x \in \mathcal X} p(x).$$

For a continuous RV, which can take on a whole *continuum* of values, the probability that the RV is any one specific value is 0. Instead, we talk about the **probability density function** (pdf), $f(x)$. The pdf exists to be integrated, and the values you get after integration are probabilities. For example, to calculate what is the probability that a continuous RV takes on values between $a$ and $b$, we write:

$$ \mathbb P[a < X < b] = \int_{a}^b f(x) dx.$$

The use of the words **mass** and **density** are purposeful here, and you should rely on your intuition from physics to intepret what they mean in a probability context. 

We will often conflate RVs with the probability function, as the two are equivalent. A RV will have a probability mass or density function and a probability mass/density function will conversely determine a RV. Therefore, we will often discuss e.g. a Gaussian *RV* and the Gaussian *distribution* and these essentially mean the same thing for discussion.

Some distributions occur so much that they are given special names and symbols. Here are a few:

* $\text{Unif}(a,b)$ - the [continuous uniform distribution](https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)). The probability is uniformly spread out throughout the interval $[a, b]$. It's pdf is: 

$$f_{\text{Unif}(a,b)}(x) = \begin{cases} \frac{1}{b-a} & x \in [a,b] \\ 0 & \text{ otherwise} \end{cases}.$$

* $\text{Unif}(\mathcal X)$ - the [discrete uniform distribution](https://en.wikipedia.org/wiki/Discrete_uniform_distribution). Here, $\mathcal X$ is a set of possible values, necessarily finite. The probability that a RV with this distribution equals any particlar value inside $\mathcal X$ is the same for all $x \in \mathcal X$. If $|\mathcal X|$ denotes how many elements are inside $\mathcal X$, then the pmf of this distribution is:

$$ p_{\text{Unif}(\mathcal X)}(x) = \begin{cases} \frac{1}{|\mathcal X|} & x \in \mathcal X \\ 0 &  \text{ otherwise} \end{cases}.$$

* $\mathcal N(\mu, \sigma^2)$ - the [Gaussian or Normal distribution](https://en.wikipedia.org/wiki/Normal_distribution). This is the bell curve you are familiar with. It is a continuous distribution that occurs frequently in nature due in no small part to something called the [Central Limit Theorem](https://en.wikipedia.org/wiki/Central_limit_theorem), which will be introduced later. Normal distributions are also easy to work with, so they are frequently used in analysis. The distribution is completely described by two numbers: the mean value $\mu$ describing the center of the bell curve, and the variance $\sigma^2$, the square-root of which $\sigma$ describes the width of the bell curve. The Normal pdf is given by

$$ f_{\mathcal N(\mu, \sigma^2)}(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left[-\frac{(x - \mu)^2}{2\sigma^2}\right].$$

* $\text{Pois}(\lambda)$ - the [Poisson distribution](https://en.wikipedia.org/wiki/Poisson_distribution). Often used to model the number the occurances of a rare event. It is parameterized by a single number, $\lambda$ which describes the rate of such events. It is a discrete distribution with pmf

$$ p_\text{Pois$(\lambda)$}(k) = \frac{\lambda^k e^{-\lambda}}{k!}.$$

* $\text{Binom}(n, p)$ - the [Binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution). This discrete distribution model the number of "success" that can occur after $n$ trials, where each trial is successful with a probability of $p$. It's pmf is:

$$ p_\text{Binom$(n,p)$}(k) = \binom{n}{k} p^k (1-p)^{n-k},$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; where $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ is the [binomial coefficient](https://en.wikipedia.org/wiki/Binomial_coefficient).

We'll encounter other distributions / RVs throughout the semester.

## Summary statistics

Given a RV $X$ and its distribution function, we can calculate some summary statistics, which are meant to capture properties of the distribution function. For example, the **expected value** of $X$, denoted $\mathbb E[X]$ describes the average of all the possible values $X$ can take, weighted by distribution function. For discrete RVs:
$$ \mathbb E[X] = \sum_{x \in \mathcal X} x p(x).$$

For continuous RVs:

$$ \mathbb E[X] = \int_{x \in \mathcal X} x f(x) dx.$$

In both cases, we use $\mathcal X$ to denote the domain or support of the density function. 

We can similarly define expected value of some function of $X$. If $Z = g(X)$ is another random variable defined as a function of $X$, we can use $X$'s density function to calculate the $\mathbb E[Z]$ (for the continouous case):

$$ \begin{aligned}
   \mathbb E[Z] &= \mathbb E[g(X)], \\
   &= \int_{x \in \mathcal X} g(x) f(x) dx.
   \end{aligned}$$
   
For the discrete case, you get something similar
$$\mathbb E[Z] = \sum_{x \in \mathcal X} g(x) p(x).$$

The **variance** of $X$, denoted $\text{Var}[X]$, describes the spread of the distribution. By definition, it is an expected value:

$$\begin{aligned}
\text{Var}[X] 
&= \mathbb E[ (X - \mathbb E[X])^2 ] \\
&= \mathbb E[X^2] - (\mathbb E[X])^2
\end{aligned}$$

The variance is also known as the 2nd central moment of the distribution. In general, the **n**-th central moment
is defined as

$$m_n = \mathbb E[(X - \mathbb E[X])^n].$$

Collectively, the central moments of the distribution are enough to define the entire distribution, and each moment describes some aspect of the distribution. The mean and variance are examples of this, while higher momements include skewness (3rd moment) and kurtosis (4th moment).


# Flipping a biased coin
---

Here we explore a few of these ideas through one of the most basic random process: flipping a coin. This one coin flip will result in either a "Heads" ($H$) or "Tails" ($T$). We'll assume that the outcome of a coin flip results in a $H$ with some probability $p$, which conversely means that we get a $T$ with probability $1-p$. If $p = 0.5$, i.e. both $H$ and $T$ are equally likely to occur, then the coin is called **unbiased**, otherwise, we call the coin **biased**.


Some times, the outcome of a coin flip is called a *Bernoulli trial*, and besides "Heads" vs. "Tails", Bernoulli trials are said to produce "Success" vs. "Failure" outcomes as well. If instead, you assign the number $1$ to a heads (or success) outcome, and a $0$ to a tails (or failure), then the outcome of a Bernoulli trial is sometimes called a **Bernoulli Random Variable**. If $X$ is a Bernoulli random variable, we will sometimes write 

$$ X \sim \text{Bern}(p)$$

Again, the parameter $p$ describes the probability of a success. Throghout this note, I'll use "Heads" and "Success" synonymously, and the same with Tails and "Failure".

In this section, we will simulated the flipping of a coin and derive other random variables from such flips. To do this first, we'll need the ability to generate some random variables.

## Coin Flipping simulations

Drawing (sampling, generating) a Bernoulli RV, is fairly simple, given the ability to draw a uniform random variable $\text{Unif}(0,1)$. Sample $u$ as a sample of the uniform RV $U \sim \text{Unif}(0,1)$. If $u < p$, then set $x = 1$ (success). Otherwise, set $x = 0$. Then $x$ is a sample of a Bernoulli RV with success probability $p$.

```python
import random
def sample_bern(rng, p=0.5):
    u = rng.random()
    
    if u < p:
        return 1
    else:
        return 0
```

Once we define this, we can sample Bernoulli RVs. For example, by setting the passed-in $p$ value to a small number, we would expect that most of the time we get a failure (0).

```python
x = sample_bern(rng, 0.2)
print(x)
```

Try running the above cell lots of times and see how often you get a 1 (i.e. a "Head"). If $p = 0.2$, then 2 in about 10 of the flips should result in Heads. 

Instead of running this lots of times, we'll just use a list comprehension to sample 100 Bernoulli RV and save them to a list called `samples`:

```python
samples = [sample_bern(rng, 0.2) for i in range(100)]
print(samples)
```

That is: counting how many successes are within a set of $n$ Bernoulli RVs is a Random variable.

This random variable is called the **Binomial Random Variable**, and it counts the number of heads after $n$ coin flips with a coin with bias $p$, and we can denote such an RV as $Y \sim \text{Binom}(n, p)$.

If $X_1, X_2$, ..., $X_n$ are indepedented Bernoulli RVs corresponding to each of the $n$ flips, we write:

$$ X_i \sim \text{Bern}(p), \quad i = 1, 2, ... n, \quad iid.$$

where "iid." stands for identically and inpendently distributed -- each $X_i$ is generated from the same coin, but the result of one flip doesn't impact the results of any other flip in any way. We can then define a Binomial RV like we did above: just sum the $x_i$ together.

$$Y = X_1 + X_2 + ... + X_n$$

This lets us write a Binomial RNG:

```python
def sample_binom(rng, n, p = 0.5):
    # get n Bernoulli rvs
    samples = [sample_bern(rng, p) for i in range(n)]
    
    # Sum them to get a Binomial RV
    return np.sum(samples)
```

Using this, we can sample from the binomial distribution. Here we'll sample 10 such Binomial RVs

```python
binom_samples = [sample_binom(rng, 100, 0.2) for _ in range(10)]
```

And we can plot the histogram of samples:

```python
# rwidth adds some space between the plots
plt.hist(binom_samples, rwidth = 0.9)
```

Of course, we can sample more than 50 RVs. Below we plot the histograms obtained after sampling 100, 1000, and 10000 values from $\text{Binom}(n, p)$.

```python
sample_sizes = [100, 1000, 10000, 100000]
plt.figure(figsize = (10, 3))
for i in range(len(sample_sizes)):
    binom_samples = [sample_binom(rng, 100, 0.2) for _ in range(sample_sizes[i])]    
    plt.subplot(1,len(sample_sizes), i + 1)
    plt.hist(binom_samples, bins = 10, rwidth = 0.9)
    
    plt.title('m = %d'%sample_sizes[i])
    
plt.tight_layout()
```

    * If you use subplots, then you should always call `plt.tight_layout` at the end to deal with layout issues.

As we get more and more samples, the normalized histograms start looking like the actual probability mass function of the Binomial distribution, which again is:


 $$ p_\text{Binom$(n,p)$}(k) = \binom{n}{k} p^k (1-p)^{n-k},$$
 
 To see this, we'll implement this function. We'll use the `math.comb` function, which is new in Python 3.8 (one of the newer version of Python).

```python
import math

def binom_pmf(k, n, p):
    return math.comb(n,k)*(p**k)*(1-p)**(n-k)
```

Once defined, we can evaluate it at various values of $k$, which corresponds to the number of heads after $n$ coin-flips. Evaluating the pmf at different $k$ values provides the exact probability of seeing $k$ heads. First we determine which $k$ values to evaluate and plot. We use the `np.arange` function, below to define these values, which are just the integers between 0 and 20.

```python
k_to_plot = np.arange(0, 35, 1)
print(k_to_plot)
```

Now we evaluate `binom_pmf` for all `k` inside this array using a list comprehension.

```python
pmf_vals = [binom_pmf(k, 100, 0.2) for k in k_to_plot]
```

Now we can plot the pmf function vs. $k$. We'll use the `plt.plot` function that takes in a list of input values and a corresponding list of output values. We'll also label our axis

```python
plt.plot(k_to_plot, pmf_vals)
plt.xlabel('k')
plt.ylabel('Probability')
```

We can overlay this PMF with each of the normalized histograms to see how close they get to approximating it. To do that, we just add the call to `plt.plot` after we plot the histogram.

```python
sample_sizes = [100, 200, 400, 800, 1600]
plt.figure(figsize = (10, 3))
for i in range(len(sample_sizes)):
    binom_samples = [sample_binom(rng, 100, 0.2) for _ in range(sample_sizes[i])] 
    plt.subplot(1,len(sample_sizes), i + 1)
    plt.hist(binom_samples, rwidth = 0.9, density = True)
    plt.title('m = %d'%sample_sizes[i])
    
    # Add the pmf plot and axes labels
    plt.plot(k_to_plot, pmf_vals)
    plt.xlabel('k')
    plt.ylabel('Probability')
    
plt.tight_layout()
```

In general, as the number of samples increases, the (normalized) histogram should start looking more like the pmf (for discrete distribution) or the pdf (for continuous distributions). This is the perspective that is taken in frequentist statistics - probabilties are viewed as histograms for a very large sample size.

## Sample Average

We can be a little bit more quantatative here by calculating some sample statistics and comparing them with the summary statistics of the true distrbution. For example, we can calculate is the average of the sampled values. Below we'll do this for different number of sample sizes.

```python
sample_sizes = np.arange(100, 5000, 100)

# Pre-allocate the array that will store the average of the sampled values.
sample_means = np.zeros(len(sample_sizes))

for i in range(len(sample_sizes)):
    binom_samples = [sample_binom(rng, 100, 0.2) for _ in range(sample_sizes[i])] 
    sample_means[i] = np.mean(binom_samples)
```

Once we have this, we can plot what the sample mean is vs. sample size, again using the` plt.plot function`.

```python
plt.plot(sample_sizes, sample_means)
plt.xlabel('Number of samples')
plt.ylabel('Average of samples \nof Binom. RV')
```

We note a few things here. First, we see that the plot isn't smooth, which suggests a random or noisy component to the average value we are calculating. In fact, if you were to run the sampling code again and plot the sample means, you would get a different plot:

```python
# Pre-allocate the array that will store the average of the sampled values.
sample_means = np.zeros(len(sample_sizes))

for i in range(len(sample_sizes)):
    binom_samples = [sample_binom(rng, 100, 0.2) for _ in range(sample_sizes[i])] 
    sample_means[i] = np.mean(binom_samples)
    
plt.plot(sample_sizes, sample_means)
plt.xlabel('Number of samples')
plt.ylabel('Average of samples \nof Binom. RV')
plt.title('Re-running the sampling')
```

This shows that the average of the sampled values is random quantity. More precisely, the average of the sampled values is itself a sample of a random variable called the **sample average**. In symbols, if $X_1, X_2, ..., X_m$ are $m$ iid RVs, then we define their **sample average**, $Z_m$,as:

$$ Z_m = \frac{1}{m} \cdot(X_1 + X_2 + \cdots + X_m).$$

Because $Z_m$ is a sum of RVs, it is itself a RV. If we have samples $x_1, x_2, ..., x_m$ of the RVs $X_i$, then their average

$$ z_m = \frac{1}{m} \cdot(x_1 + x_2 + \cdots + x_m).$$

is a sample of $Z$.

Another thing we notice about the above plots is that the average of the sample varies more when the number of samples is small. As the number of samples increases the value tends to (or **converges**) to a specific number. Here it looks like that number is around 20. This is related to the expected value and variance of the RV $Z_m$. In fact we have the following.

---

**Proposition** Suppose $X_1, ... , X_m$ are iid RVs with expected value $\mu$ and variance $\sigma$. Let $Z_m$ be defined as above. Then the expected value $\mathbb E[Z_m] = \mu$ and variance $\text{Var}[Z_m] = \frac{\sigma^2}{m}$.

---

You may be asked to prove this. The proof depends on the linearity of expected values and variances we discussed in class.

Applying this to our situation, each of the $X_i \sim \text{Binom}(n,p)$. Therefore,

$$\mu = \mathbb E[X_i] = np$$

and

$$\sigma^2 = \text{Var}[X_i] = np(1-p).$$

You should prove to yourself that this is true. (Hint: you can use the fact that $X_i$ is itself a sum of Bernoulli RVs, and the above proposition).

Therefore, the expected value of $Z_i$ is also $np$, while it's variance is $\frac{np(1-p)}{m}$. When $n = 100$ and $p = 0.2$, we see that the expected value of $Z_i$ is 20, while it's variance is $\frac{16}{m}$. Therefore, if we sample $z_i$, on average it should be around 20, and the variance around this value decreases as $m$ gets bigger.


### Some things to ponder about*

There is something a bit subtle going on here that isn't necessarily appreciated the first time people think about averaging samples. When we sample the $X_i$ to get samples $x_i$ and form an average $z$, there are in fact three related concepts floating around:

1. The expected value $\mathbb E[X_i]$, which we expect the samples $x_i$ to equal, on average;
2. The limit of average of the samples $\lim_{m \rightarrow \infty} z_m = (x_1 + \cdots + x_m) / m$; and
3. The expected value $\mathbb E[Z_m]$, where $Z_m = (X_1 + \cdots + X_m) / m$.

All three numbers are equal in this setting. This should give you a moments pause to ponder:

* The formula to compute $z_m$ from the samples $x_i$ is the same as the formula for defining $Z_m$ from the RVs $X_i$. Does this necessarily mean that $z_m$ is a sample of $Z_m$? Why?

* The expected value of the sample average $Z_m$ equals the expected value of the $X_i$. This may seem obvious at first, but we had to prove it using the above proposition.

To offer a clarifying counterpoint as to why this should cause you to pause for a bit, we can simplify the situation here. In essense, we defined one random variable $Z$ in terms of others -- we had some transformation to turn the $X_i$ into $Z$. Let's simplify and abstract this a bit, by considering only one $X$ and writing $Z = g(X)$ as some abritrary transformation of $X$. Ponder the following

* If $x$ is a sample of $X$, is $g(x)$ a sample of $Z$?
* If $f_X$ is the pdf/pmf of $x$, what is the pdf/pmf of $Z$ in terms of $f_X$?
* If $\mu = \mathbb E[X]$, is $\mathbb E[Z] = g(\mu)$? That is, is the expected value of the transformation equal to the transformation of the expected value?

Consider writing some simulations like we did above. Instead of computing the average of $X_i$, what happens if you try something else?

## The Normal  approximation to the Binomial RV.

The pmf of the Binomial distribution is pretty messy to deal with, which means calculating summary statistics or conditional distributions is quite difficult to do, at least before we had computers. For example, we can ask what is the probability that a biased coin flipped 100 times comes up heads at least 50 times. In symbols, what is $$\mathbb P[X \geq 50], \quad X\sim\text{Binom}(n,p)?$$

By definition, we have
$$\mathbb P[X \geq 50] = \sum_{k=50}^{100} p_\text{Binom}(k)$$

Expanding the pmf, we get the uninsightful expression

$$\mathbb P[X \geq 50] = \sum_{k=50}^{100} \binom{n}{k} p^k (1-p)^{n-k}$$

However uninsightful, using computers we can nevertheless calculate this sum for specific values of $n$ and $p$. We'll use the `scipy.stats.binom.pmf` to evaluate the terms of the sum. We'll calculate each of the terms in the sum firs, then sum them up.

```python
# The numbers from 50 to 100, inclusive
k_vals = np.arange(50, 101)
print(k_vals)
```

```python
from scipy.stats import binom
# Calculates the binomial pmf values for each k in k_vals
pmf_vals = binom.pmf(k_vals, n = 100, p = 0.45)
```

We can plot the pmf vals just to visualize it.

```python
plt.plot(k_vals, pmf_vals)
```

We can also sum these values to get the probability that we wanted.

```python
prob_at_least_50 = np.sum(pmf_vals)
print("Probability of at least 50 heads = %f"%prob_at_least_50)
```

Similarly, if we wanted to calculate the probabiliy that the number of heads was at least 20:

```python
k_vals = np.arange(20, 101)
pmf_vals = binom.pmf(k_vals, n = 100, p = 0.45)
plt.plot(k_vals, pmf_vals)

prob_at_least_20 = np.sum(pmf_vals)
print("Probability of at least 20 heads = %f"%prob_at_least_20)
```

Looking at the above, we see that this pmf looks a lot like the pdf of the Normal distribution, the bell curve. In fact, recall the previously presented mean and variance values for the Binomial distribution, namely:

$$ \mu = \mathbb E[X] = np,$$

$$ \sigma^2 = Var[X] = np(1-p).$$

We could use these parameters as the same parameters to the Normal distribtuion pdf:

$$f_{\mathcal N(\mu, \sigma^2)}(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left[-\frac{(x - \mu)^2}{2\sigma^2}\right].$$

Other than observing the shape of the curve above, we had no reason to do so, but we can plug these values into the Normal pdf nonetheless. Below we do this for a continous set of $x$ values (really 200 uniformly-spaced values between 20 and 100). We also calculate the parameters defined above.

```python
x_vals = np.linspace(20, 100, 200)

# Compute the mean and variance of the Binomial distribution to plug into the normal distribution.
n = 100
p = 0.45

mu = n*p
sigma_squared = n*p*(1-p)

# Take the square root of variance -- we'll need it to use scipy's Normal pdf 
sigma = np.sqrt(sigma_squared)
```

Now we use the Normal pdf value inside `scipy.stats.normal` and evaluate the pdf function at each of the values inside `x_vals`.

```python
from scipy.stats import norm

pdf_vals = norm.pdf(x_vals, mu, sigma)
```

Now we can plot both the normal pdf values and the binomial pmf values. We'll plot the pdf values with dots using the `linewidth=0` and `marker = '.'` arguments to `plt.plot`. We'll also add a legend.

```python
plt.plot(k_vals, pmf_vals)
plt.plot(x_vals, pdf_vals, linewidth=0, marker ='.')

plt.legend(['Binomial PMF', 'Normal PDF'])
```

We see that the normal pdf is a very good approximation to the Binomial pmf.

That we can approximate the Binomial pmf with the normal pdf is somewhat profound, or at least leads to some profound applications. One of the first things to note is that a Binomial RV is a discrete RV - it can only take on integer values from 0 to $n$. However a Normal RV is continouous. It can take on values between negative infinity and positive infinity. 

The ability to switch from a discrete object to continuous is fairly powerful, and is something you will see a lot throuhgout science, including mechanics, kinetics, statistical. The reason this is powerful is that it allows us to use calculus to study your system. For example, this specific case of the binomial distribution being approximated by the Normal distribution allows us to go from thinking about discrete steps of a particle to the continuous theory of diffusion (we'll go into this later).

# The Central Limit Theorem
---

Of course, we do not get a normal distribution approximation just by calculating the mean and variance of any RV and plugging them into the normal PDF. For example, let's consider the mean and variance of a Uniform RV, $X \sim U(0,1)$.

Here (after some calculation):

$$\mathbb E[X] = \frac{1}{2},$$

$$\text{Var}[X] = \frac{1}{12},$$

The pmf of $X$ is fairly simple:

$$p(x) = \begin{cases} 1 &  0 \leq x \leq 1 \\ 0 &  \text{ otherwise} \end{cases}.$$

Plotting this and the corresponding normal approximation obtained by plugging in the above for the mean $\mu$ and variance $\sigma^2$ parameters, we get the following.

```python
# Lets plot both pdf between -1, 2 (100 equal spaced points)
x_vals = np.linspace(-1, 2, 200)

# The uniform pdf
from scipy.stats import uniform
unif_pdf_vals = uniform.pdf(x_vals, 0, 1)
plt.plot(x_vals, unif_pdf_vals)

# The normal pdf
mu = 0.5
sigma = 1/12
norm_pdf_vals = norm.pdf(x_vals, mu, sigma)
plt.plot(x_vals, norm_pdf_vals)
```

These don't look the same at all.

What makes the Normal distribution work for the Binomal RV is the fact that it is actually a sum of iid Bernoulli RVs. In fact, we can generalize this sums of **any** RV. This, in essense, is the **central limit theorem** we discussed in class and can be paraphrased as following (with some caveats): *Sums of RVs tend to behave like Normal RVs*.

To see this, let us calculate the sum of samples of uniform RVs lots of times, and look at the distribution of such sums. For example, we can consider 1000 sums of 200 samples of a uniform RVS. Let's calculate these 1000 sums and plot their distribution. There are lots of ways to do this, but this time we'll use the `numpy` RNG. We'll sample a $1000 \times 200$ matrix where each entry is a sample from $U(0,1)$.

```python
unif_values = rng.uniform(0, 1, [1000, 200])
print(unif_values.shape)
```

Then we'll sum all the rows

```python
row_sums = np.sum(unif_values, axis = 1)
print(row_sums.shape)
```

Now we can plot the histogram of the row sums

```python
plt.hist(row_sums, bins = 20, rwidth = 0.98, density = True)
```

This looks pretty Normally distributed. To verify this, let's calculate the mean and variance. We'll use sample means and variances using `np.mean` and `np.var`.

```python
mu = np.mean(row_sums)
sigma_squared = np.var(row_sums)

print('mu = %f, sigma_squared = %f'%(mu, sigma_squared))
```

Now we'll plot the normal pdf with these mean and variances.

```python
# The histrogram of row sums
plt.hist(row_sums, bins = 20, rwidth = 0.98, density = True)

# The normal pdf
x_vals = np.linspace(85, 115, 200)
norm_pdf_vals = norm.pdf(x_vals, mu, np.sqrt(sigma_squared))
plt.plot(x_vals, norm_pdf_vals)
```

This looks like a good fit. It would appear that the central limit is true, but you should try to explore its limitation, in particular the number of uniform RVs we added together to get a sum of Uniform RVs.
